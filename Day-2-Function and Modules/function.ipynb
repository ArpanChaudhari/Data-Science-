### FUNCTION

```python
# Create user-defined function with parameter
def greet(user):
    print(f"Hello {user}")

# call function
greet("Arpan")
```

```python
# Function with Multiple parameter
def addition(n1,n2) :
    print(n1+n2)

addition(24,45)
```

Call by value :- immutable data types are called by value

```python
def increment(n) :
    n+=1
    print("In function :",n)

n=10
print("Before function :",n)
increment(n)
print("After function :",n)
```

Call by reference :- all mutable data type values are called by reference

```python
def addEleList(lst) :
    lst.append(1234)
    print("In function :",lst)

lst = [1,2,3,4]
print("Before function :",lst)
addEleList(lst)
print("After function :",lst)
```

Return statement

```python
def add(a,b):
    return a+b

print(add(4,6))
```

### Types of Function Arguments:

1. Default Argument

```python
def add(a,b=10) :
    print(a,"+",b,"=",a+b)

add(10)
add(10,45)
```

2. Keyword Argument

```python
# Keyword arguments (also known as named arguments) allow you to pass values to a function by explicitly specifying the parameter name followed by an equal sign (parameter=value).
def users(name,age,addr):
    print(name,age,addr)

users("Arpan",21,"Ahmedabad")
users(age=21,name="Arpan",addr="Ahmedabad")
users(addr="Ahmedabad",age=21,name="Arpan")
```

3. Positional Argument

```python
def user(name,age):
    print(name,age)

user("Arpan",21)
user(21,"Arpan")
```

4.Arbitary Argument

i) *args :- Collects multiple positional arguments into a tuple ()

```python
def fun(*args) :
    return sum(args)

print(fun(45,36,35))
```

```python
def multiply(*args) :
    result=1
    for i in args :
        result*=i
        return result
    
print(multiply(4,5,6))

```

ii) **Kwargs :- Collects keyword arguments into a dictionary {}

```python
def myFun(**kwargs) :
    for k,v in kwargs.items() :
        print(k,':',v)

myFun(name="Arpan",age=21,address="Ahmedabad")
```

### Doc String
Docstrings (Documentation Strings) are special strings used to document Python code.accessed at runtime using __doc__ or help().

Declared using triple quotes (' ' ' or " " ").

```python
def adds(s1,s2):
    '''Function returns concatenated string'''
    s = s1+s2
    return s

print("Using __doc__:")
print(adds.__doc__)

print("Using help():")
help(adds)
```

### YIELD vs RETURN

Yield is generally used to convert a regular Python function into a generator.

Return is generally used for the end of the execution and “returns” the result to the caller statement.

```python
# return
def fun1(n) :
    for i in range (1,n) :
        return i**2
    
print(fun1(5))
```

```python
# yield

def fun2(n) :
    for i in range (1,n) :
        yield i**2

print(list(fun2(5)))
```
