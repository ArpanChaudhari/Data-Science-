1.Lambda Function

```python
add= lambda x:x+10
print(add(45))

sq = lambda x:x**2 
print(sq(16))
```

2. map

```python
# map(function , iterable)

lst = [1,2,3,4,5]

sqrs=list(map(lambda x:x**2 , lst))
print(sqrs)
```

3.filter

```python
even=list(filter(lambda x:x%2 ==0 , lst))
print(even)
```

4.reduce

```python
from functools import reduce

sum=reduce(lambda a,b :a+b ,lst)
print(sum)
```

5.sorted

```python
list=[2,4,1,3,9,5]
res=sorted(list)
print(res)
```
