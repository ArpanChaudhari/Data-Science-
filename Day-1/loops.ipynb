### WHILE LOOP

```python
i=1 
while i<5 :
    print(i)
    # i=i+1
    i+=1
```

While loop with else block

```python
i=1 
while i<5 :
    print(i)
    i+=1
else :
    print("Loops end")
```

```python
# infinite loop
i=1
while True :
    print(i)
    i+=1
```

### FOR LOOP

With range(start , end - 1 ) function

```python
for i in range(1,5) :
    print(i)
```

Iteration over list, tuple, set, string, dict 

```python
print("List Iteration")
lst = [ 1,2,4,5,6]
for i in lst :
    print(i , end=' ')
# end=' ' replaces that newline with a single space. This allows you to print multiple outputs on the same line
print()

for i  in range(len(lst)) :
    print(lst[i] , end=' ')
print()

for i, e in enumerate(lst) :
    print(i,e)
```

```python
print("Tuple Iteration")

tup=(2,4,6,7,8,9)
for i in tup :
    print(i, end=' ')
print()

for i in range(len(tup)) :
    print(tup[i], end=' ')
print()

for i,e in enumerate(tup) :
    print(i, e)
```

```python
print("Set Iteration")

set={11,13,23,22,21,11,13,14,22,23,11,22,11}

for i in set :
    print(i, end=' ')

# Set does not  support indexing like list and tuple
```

```python
print("Dictionary Iteration")

dict={1:11,2:22,3:33,4:44,5:55}

print("print keys")
for i in dict.keys() :
    print(i , end=' ')

print("\nprint values")
for i in dict.values() :
    print(i, end=' ')

print("\nprint item")
for i in dict.items() :
    print(i, end=' ')

print("\nprint keys with values")
for i in dict.keys() :
    print(i,":",dict[i])
```

```python
print("String Iteration")

str="Arpan"

for i in str :
    print(i, end=',')
print()

for i in range(len(str)) :
    print(str[i], end=',')
```

### NESTED LOOPS

```python
for i in range(1,6) :
    for j in range(i):
        print('*', end='')
    print()
```

```python
rows = 5
for i in range(1, rows + 1):
    # Print leading spaces
    print(" " * (rows - i), end="")
    # Print stars
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
n=5
for i in range(1,n+1) :
    for j in range(n-i) :
        print(" ", end='')
    for k in range(1,2*i) :
        print("*",end='')
    print()
```

### LOOPING CONTROL STATEMENT 

```python
for i in range(1,10):
    if i % 2 == 0 :
        continue
    print(i)
```

```python
for i in range(1,15):
    if i%5 == 0 :
        break
    print(i)
```
