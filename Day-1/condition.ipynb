### IF STATEMENT

```python
i=int(input("Enter the student's marks:"))
if i>5 : 
    print('i is greater than 5..')
```

### IF-ELSE STATEMENT

```python
i=int(input("Enter the student's marks:"))
if i>5 : 
    print('i is greater than 5..')
else :
    print('i is less than 5')
```

### IF-ELIF-ELSE STATEMENT 

```python
i=int(input("Enter the student's marks:"))
if i<5 :
    print('i is less than 5..')
elif i==5 :
    print('i is equal to 5..')
else :
    print('i is greater than 5..')
```

### NESTED-IF STATEMENT

```python
age=int(input("Enter the student's marks:"))
has_license = True

if age >= 18 :
    if has_license :
        print('You can drive.')
    else :
        print('you need a license.')
else : 
    print("You are too young to drive.")
```

### SHORT HAND IF STATEMENT

```python
# Short hand if statement (if condition: statement)

i=10
if i>15 : print('i is less than 15..')
```

```python
# short hand if-else statement - Ternary Operator (val1 if cond else val2)

i=10
print(True) if i==10 else print(False)
```

```python
# chained ternary (v1 if c1 else v2 if c2 else v3)

a=4
b=5
print("A") if a > b else print("=") if a == b else print("B")

# a>b -> A, a=b -> =, a<b -> B
```

### MATCH CASE STATEMENT

```python
person=str(input("Enter Number:"))

match person :
    case "A":
        print("Hello, A!")
    case "B":
        print("Hello, B!")
    case _:  
        #_ (Wildcard): A default catch-all pattern, similar to a "default" in other languages' switch statements.
        print(f"Hello, {person}!")
```

WITH OR OPERATOR

```python
num=int(input("Enter number:"))

match num :
    case 10 | 20 | 30 :
        print(f"matched: {num}")
    case _:
        print(f"No match found , number is: {num}")
```

WITH IF STATEMENT

```python
price=int(input("Enter a price:"))

match price :
    case _ if price < 10 :
        print(f":low-cose - {price}")
    case _ if 10 <= price < 50 :
        print(f"Moderate - {price}")
    case _ :
        print(f"Luxury - {price}")
        
```

MATCH THE SEQUENCE PATTERN

```python
# data=input("Enter element of list:").split()

data=list(map(int,input("Enter element of list:").split() ))

match data:
    case [x,y] :
        print(f"Two Element list : {x},{y}")
    case [x,y,z] :
        print(f"Three element list : {x},{y},{z}")
    case _ :
        print("Unknown data format")
```
