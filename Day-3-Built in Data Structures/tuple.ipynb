# Tuple ()

### Key Properties:
* Allows duplicates
* Ordered
* Immutable (cannot change after creation)
* Indexing & slicing supported

```python
# length and type

t = (1,2,3,4,5)
print(len(t))
print(type(t))
```

```python
# max,min,sum

t = (2,3,4,5,6,3,6)

print(max(t))
print(min(t))
print(sum(t))
```

```python
# Indexing and slicing from the tuple

t = (2,3,4,5,6,7,8)
print(t[3])
print(t[-2])
print(t[1:4])
print(t[-4:-1])
print(t[-4:-5]) # t[start : end : step] hear start:-4 end:-5
# empty tuple becouse it moves left -> right it stop before end
print(t[-4:-7:-1])  # start > end -> use negative step

print(t[1:6:2])
```

```python
# Count and Index

p=(1,2,3,4,2,3,4,5,2,3,4,2,3,1,3,2)

print(p.count(2))

#index(value,start,end)
print(p.index(2,5,9))
print(p.index(2))
```
