# List []

### Key Properties:
* Allows duplicates
* Ordered (maintains insertion order)
* Mutable (can modify elements)
* Indexing & slicing supported

```python
s=input()
```

```python
print(s)
```

```python
l=s.split()
print(l)
```

```python
i=[int(i) for i in s.split()]
print(i)

# int(i) - convert each element of s.split() into integer and assign in i[] list
```

```python
i.extend((6,7,8))
print(i)
```

```python
i.extend([9,10,11])
print(i)
```

```python
i.extend({12,13,14})
print(i)
```

```python
i.insert(1,[0,0])
print(i)
```

### List Comprehension

```python
d=[i**2 for i in range(10)]
print(d)
```

```python
even = [i for i in range(1,21) if i%2 ==0]
print(even)
```

```python
odd = [i for i in range(1,21) if i%2 !=0]
print(odd)
```

```python
# 3. Matrix generation using list comprehension

mat=[ [j for j in range(3)] for i in range(3)]
print(mat)
```

```python
letters = [ char for char in "Arpan Chaudhari"]
print(letters)
```
