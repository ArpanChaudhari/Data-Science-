# Dictionary 

### Key Properties:

* Keys do NOT allow duplicates (values can repeat)
* Mutable → Not allowed as key
* Ordered (Python 3.7+)
* Mutable
* Uses key-value pairs
* No indexing (access via keys)

```python
d={1:"Arpan" , 2:"Ayush" ,3:"Yash"}
print(d)
```

```python
sub={"os" , "db" ,"DSA"}
d={'name':"Arpan" , 'age':22 ,'address':"Ahmedabasd",'marks':[99,100] , 'subject':sub}
print(d)
```


Create Dictionary with dict() method

```python
d=[(1,"Arpan"),(2,"Ayush"),(3,"Yash")]
t=dict(d)
print(t)
```

```python
sub={"os" , "db" ,"DS"}
a=dict(sub)
print(a)
```

```python
d={'name' : "sne" , 'age':22,'add':["Ahmedabad","Surat"]}
print(d["add"])
```

```python
# If we ean access some key but not present in dictionary 
print(d.get("marks"))  # its give output:- None
print(d["marks"])  # error
```

```python
print(d.items())
print(d.keys())
print(d.values())
```

Loop through the Dictionary

```python
for i in d.keys() :
    print(i,":",d[i])
```

```python
for i in d.items() :
    print(i)
```

```python
# d.items() returns key–value pairs as tuples
for i in d.items() :
    print(i[0],":",i[1])
```

```python
for i in d.values() :
    print(i)
```

```python
print(id(d["name"]))
# id - Return the identity of an object.
```

```python
d["Name"] = d["name"]
print(d)
```

```python
print(id(d['name']))
print(id(d['Name']))

''' Both lines will print the same memory address
Because : 
  You are not copying the value
  You are assigning the same object reference
So both keys point to the same object in memory'''
```

```python
d['Age'] = d['age']
print(d)
```

```python
print(id(d['age']))
print(id(d['Age']))
```

```python
d['Age'] = 25
```

```python
print(id(d['age']))
print(id(d['Age']))
```

```python
d["Address"] = d.pop('add')
print(d)
```

Update() function

```python
d1 = {'name': 'Sne'}
d2 = {'age': 21}
d3 = {'addr': 'Kop'}

d1.update({'name' : "Arpan"})
d1.update(d2)
d1.update(d3)

print(d1)
```

```python
d = {'name': 'm'}

d.setdefault("name" ,'s')  # 'name' already exists - So value does NOT change
d.setdefault('add','Ahmedabad') # 'addr' does NOT exist - So it gets added with value 'Kop'
d.setdefault('sub') # 'sub' does NOT exist  - No default value given → default is None

print(d)
```

Removing items from the dictionary

```python
d={'name' : "Arpan" , 'age':22,'add':["Ahmedabad","Surat"]}
print(d)

d.pop('name')
print(d)

del d['age']
print(d)

d.clear()
print(d)

del d
print(d)
```
