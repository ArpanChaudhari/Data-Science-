# Set {}

### Key properties of sets
* No duplicates allowed
* Unordered
* Mutable

```python
s={1,2,1,2,3,4,3,5,2,3,5,7}
print(s)
```

```python
s={1,False,True,12,"Arpan",0}
print(s)

# True = 1  & False = 0
```

```python
print(len(s))
```

```python
print({1,(2,3,4,5)})
```

```python
print(s)
s.add(23)
print(s)
```

```python
s.add(14)
print(s)
```

```python
s = {1,2,3,4}
l = ['a','s','d','f']
t = (2.2, 3.4, 5.6)

d={0}
d.update(("Arpan","Ayush"))
print(d)

d.update(s)
print(d)

d.update(l)
print(d)

d.update(t)
print(d)
```

```python
s={'a','b','c','s'}
print(s)
```

```python
u1={1,2,3,4}
u2={'a','b'}
print(u1,u2)
print(u1.union(u2))

u3=[5,6,7,8]

print(u1.union(u2,u3))
print(u1 | u2)
print(u1|u2|u3)
```

```python
n=u1.union(u2)

n.remove(3)
print(n)

n.pop()
print(n)
```

```python
n.remove(1)
```

```python
n.discard(1)
print(n)
```

```python
n.clear()
print(n)
```

```python
del n
print(n)
```

```python
s = {1,2,4,3,5,7,4,6,8,4,2,5,3,6}
print(s)

print(len(s))
print(sum(s))
```

```python
f=frozenset(s) # Immutable
print(f)
```
