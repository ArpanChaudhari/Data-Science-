```python
x=5
y="Arpan"

try :
    x=x+y 
except TypeError :
    print("Error: Cannot add an int and a str")
```

# Raising User-defined Exception

```python
def fun(name) :
    try :
        if name == "Arpan" :
            print("hello ", name)
        else :
            raise NameError(f"{name} not having access..")
    except NameError:
        print(NameError.__doc__)
        raise

fun("Arpan")
fun("Ayush")
```

User Defined Exception

```python
class MyError(Exception) :
    pass 

def check(num):
    if num < 0 :
        raise MyError("Number is negative")
    print("number is  positive")

check(4)
check(-5)
```

# Error vs Exception

Error
* Serious problem
* Usually cannot be handled
* Caused by system or Python itself

```python
# print(10/0) # ZeroDivisionError
# import abcxyz   # ModuleNotFoundError
# These are errors during execution

try :
    = 10
except SyntaxError :
    print(SyntaxError.__doc__)
```

Exception
* Can be handled using try–except
* Happens due to wrong input or logic
* Program can continue after handling

```python
try:
    num = int("abc")
except ValueError:
    print("Invalid input")
```
