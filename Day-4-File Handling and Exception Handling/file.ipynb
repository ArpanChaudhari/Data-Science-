# open file

1. Open() file

```python

# open the  file geek.txt in read mode

file=open("geek.txt", "r")

if(file) :
    print("File is opened successfully")
    
file.close()
```

```python

# open the  file geek.txt in read mode

file=open("geek.txt", "r")

for each in file :
    print(each)
    
file.close()
```

read()

```python

file=open("geek.txt", "r")
print(file.read())

file.close()
```

read() and with operator

```python

with open('geek.txt','r') as file :
    print(file.read())
    
```

```python

file=open("geek.txt", "r")
print(file.read(20))

file.close()
```

readline()

```python
file=open("geek.txt", "r")
print(file.readline())
print(file.readline())
print(file.readline())

file.close()
```

```python


file=open("geek.txt", "r")

while True :
    line=file.readline()
    if not line :
        break
    print(line)
    
file.close()
```

readlines() - multiple line

```python
file=open("geek.txt", "r")

data= file.readlines()
print(data)

for i in data :
    print(i.split())

file.close()
```

write()

```python
file = open("write.txt", 'w')
file.write("This is write command")
file.write("\nWrite in the file")

file.close()
```

```python
with open("write.txt",'w') as file :
    file.write("Writing new text to the file.. \n")
    file.write("With the help of the with keyword..\n")
    file.write("Then closing the file..\n")
```

writelines()

```python
file = open("Other_file.txt",'w')
lst=["Hii\n","Hello\n","How are you doing\n"]

file.writelines(lst)

file.close()
```

append()

```python
file = open("write.txt",'a')

file.write('\nAppending to the file..\n')
file.write("Without overwriting..\n")

file.close()
```

# Close the file:

```python
file = open("geek.txt",'r')

if file :
    print("file is open successfully")

print(file.closed)
file.close()
print(file.closed)
```

Using try-finally to perform file operations :


```python
try :
    file = open("gek.txt",'r')
except FileNotFoundError :
    print(FileNotFoundError.__doc__)
finally :
    file.close()
```

```python
file = open("geek.txt",'r')
print(file)

file.close()
```
